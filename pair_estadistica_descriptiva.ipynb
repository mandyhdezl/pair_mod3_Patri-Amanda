```python
# Paso 1: Calcular los percentiles

Q1 = np.nanpercentile(df["infant_mortality"], 25)
Q2 = np.nanpercentile(df["infant_mortality"], 50)
```

```python

# Paso 2: Crear la clasificación
df["mortalidad_categoria"] = pd.cut(df["infant_mortality"],
                                    bins=[-float("inf"), Q1, Q2, float("inf")],
                                    labels=["Baja", "Moderada", "Alta"])

# Mostrar resultado
df[["country", "infant_mortality", "mortalidad_categoria"]].head()
```
