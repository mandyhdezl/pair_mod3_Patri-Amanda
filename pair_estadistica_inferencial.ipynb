# Intervalo de Confianza para la Esperanza de Vida Contexto: Trabajas en una organización de salud global y deseas estimar el intervalo 
# de confianza al 95% para la esperanza de vida promedio de cada continente. Nota: Cuidado porque puede que os salgan algunos nulos, 
# genera un código para que no salgan nulos en los intervalos de confianza.

```python
Continente=df["continent"].unique() # vemos los datos que hay en la columna de continente
```

```python
# definimos el nivel de confianza (95% en este caso)
intervalo_confianza = 0.95
```

```python
# hacemos una función para calcular el intervalo de confianza
def intervalo_confianza(data, confianza=0.95):
    data = data.dropna()  # Eliminar valores nulos
    n = len(data)  # Tamaño de la muestra
    media = np.mean(data)  # Media muestral
    sem = stats.sem(data)  # Error estándar de la media
    margen = sem * stats.t.ppf((1 + confianza) / 2., n-1)  # Cálculo del margen de error
    return media - margen, media + margen  # Retorna el intervalo inferior y superior
```

```python
# calculamos el intervalo de confianza
resultados = {}
for continente, subset in df.groupby('continent'):
    ic_vida = intervalo_confianza(subset['life_expectancy'], 0.95)
    ic_pib = intervalo_confianza(subset['PIB'], 0.90)
    resultados[continente] = {
        'IC Esperanza de Vida (95%)': ic_vida,
        'IC PIB (90%)': ic_pib
    }
```
