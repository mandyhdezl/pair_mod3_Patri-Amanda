```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("world_data_full_apply_nulos.csv", index_col=0)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
# ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente
```

```python
sns.countplot(x="continent", data=df)
plt.xticks(rotation = 90);
```

```python
Continentes = ["Asia", "Europa", "Africa", "America Central", "America del Sur", "Oceania", "America del Norte"]
```

```python
colores = ["c", "cadetblue", "mediumseagreen", "seagreen", "teal", "turquoise", "lightseagreen"]
```

```python
# empezamos preparando los datos para poder hacer el pieplot usando un groupby. En este caso tendremos que contar el número de paises por continente
df_co = df.groupby("continent")["country"].count().reset_index()
df_co
```

```python
# creamos el gráfico basándonos en el DataFrame que hemos creado en el paso anterior
plt.pie("country", labels= "continent",
        data = df, 
        autopct=  '%1.1f%%', 
        colors = colores, textprops={'fontsize': 8},startangle=90);
```
